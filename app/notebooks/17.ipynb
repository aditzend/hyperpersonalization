```python
from llama_index.readers.schema.base import Document
from llama_index import VectorStoreIndex

index = VectorStoreIndex([])
page = 0
for chunk in doc.chunks():
    
    index.insert(Document(text=chunk.to_context_text(), extra_info={}))
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("cuanto ruido hacen estas maquinas?")
print(response)
```

Estas máquinas hacen menos de 70 dB(A) de nivel de presión acústica constante en un lugar de trabajo, tanto si son enfriadas por agua como si son enfriadas por aire.
